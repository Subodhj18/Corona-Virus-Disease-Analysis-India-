```python
#Import all the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import plotly
import plotly.express as px
style.use('ggplot')
import plotly.graph_objects as go
#plt.rcParams['figure.figsize']=17,8
import cufflinks as cf
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot
import folium
```

```python
pyo.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
#reading the dataset
df = pd.read_excel('Covid cases in India.xlsx')
```

```python
df
```

```python
#Removing S. No. column for column axis=1, inplace means permanent 
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df
```

```python
#Creating new colum from two other columns
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
Total_cases_overall=df['Total Cases'].sum()
print('The total number of cases till now in India are : ',Total_cases_overall)
```

```python
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
```

```python
df
```

```python
#To display highly infected states 
df.style.background_gradient(cmap='Reds')
```

```python
Total_Active_Cases=df.groupby('Name of State / UT')['Active Cases'].sum().sort_values(ascending=False).to_frame()
```

```python
Total_Active_Cases
```

```python
Total_Active_Cases.style.background_gradient(cmap='Reds')
```

```python
df2=pd.read_csv('covid_19_data.csv',parse_dates=['Last Update'])
```

```python
df2.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
df2
```

```python
df2.query('Country=="UK"')
```

```python
df2.query('Country=="India"')
```

```python
df2.groupby('Date').sum()
```

```python
confirmed=df2.groupby('Date').sum()['Confirmed'].reset_index()
death=df2.groupby('Date').sum()['Deaths'].reset_index()
rec=df2.groupby('Date').sum()['Recovered'].reset_index()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))

fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=rec['Date'],y=rec['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```

```python
df2_confirmed=pd.read_csv('time_series_covid_19_confirmed.csv')
```

```python
df2_confirmed.rename(columns={'Country/Region':'Country'},inplace=True)
```

```python
df2_latlong=pd.merge(df2,df2_confirmed,on=['Country','Province/State'])
```

```python
fig=px.density_mapbox(df2_latlong,lat="Lat",lon="Long",hover_name="Province/State",hover_data=["Confirmed","Deaths","Recovered"],animation_frame="Date",color_continuous_scale="Portland",radius=7,zoom=0,height=700)
fig.update_layout(title='Worldwide Corona Virus Cases')
fig.update_layout(mapbox_style="open-street-map",mapbox_center_lon=0)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
```
